```python
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
iris
```

```python
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf.fit(X,y)
tree.plot_tree(clf)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
cancer_data = load_breast_cancer(as_frame=True)
print(cancer_data.DESCR)
```

```python
X, y = load_breast_cancer(return_X_y = True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clfb = tree.DecisionTreeClassifier()
clfb.fit(X_train, y_train)

prediction = clfb.predict(X_test)

for xi, yi in zip(prediction, y_test):
    print(xi,yi)
```

```python
path = clfb.cost_complexity_pruning_path(X_train, y_train)
ccp_alpha, impurities = path.ccp_alphas, path.impurities
```

```python
ccp_alpha
```

```python
plt.scatter(ccp_alpha, impurities)
```

```python
clfb.score(X_test,y_test)
```

Di seguito verr implementato il bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(tree.DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1)
```

```python
bag_clf.fit(X_train, y_train)
```

```python
bag_y_predict = bag_clf.predict(X_test)
```

```python
bag_clf.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, bag_y_predict)
```

Esempio di random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnd_clf = RandomForestClassifier(n_estimators=20, max_depth=4, bootstrap=False, n_jobs=-1)
```

```python
rnd_clf.fit(X_train, y_train)
```

```python
rnd_clf.score(X_train, y_train)
```

Esempio di boosting

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_clf = AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=4), n_estimators=500, learning_rate=1.5)
```

```python
ada_clf.fit(X_train, y_train)
```

```python
ada_clf.score(X_train, y_train)
```

```python
ada_clf.score(X_test,y_test)
```
